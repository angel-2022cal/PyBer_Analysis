```python
#import dependencies.
import matplotlib.pyplot as plt
```

```python
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
y_axis =[10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis, y_axis);
```

```python
#create a Line chart using the Object-Oriented Interfacce 

#plt.subplots() returns a tuple that contains a figure and axes objects
#fig, ax unpacks the tuple into the variables fig and ax. 
#fig references the figure
fig, ax = plt.subplots()
#ax references the axes of the figure and allows us to annotate the figure
ax.plot(x_axis, y_axis);
```

```python
# Create the plot with ax.plot()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis);
```

```python
#Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis);
```

```python
#Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label = "Boston")
#Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45.
plt.ylim(5, 45)
#Create a title 
plt.title("PyBer Fare by Month")

#Add the Legend
plt.legend();
```

```python
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 1.5, label = "Boston")

#Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45.
plt.ylim(5, 45)
#Create a title 
plt.title("PyBer Fare by Month")

#Add a grid
plt.grid()
#Add the Legend
plt.legend();
```

# 2. Bar Charts

Now, I will create and annotate vertical ans horizontal bar charts. 
Bar charts are particularly helpful when comparing datasets over time or when your line chart starts to look clustered becauses it has too many lines. 

Note: There are many benefits to using a bar chart. They are good at displaying discrete data(i.e. data based on counts) in distinct columns. 

You can use a bar chart to clearly compare two or more datasets, either by displaying them side by side, or by using a stacked bar chart. Bar charts can provide more visual appeal and informatino compared to having multiple lines on the same graph, which can mask the underlying infomration. 

#### Vertical bar chart

```python
# Create a vertical bar chart, using plt.bar() function.
plt.bar(x_axis, y_axis);
```

```python
#Create the plot.
plt.bar(x_axis, y_axis, color = "green", label = "Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")

plt.legend();
```

#### Horizontal Bar chart 

```python
plt.barh(x_axis, y_axis);
```

```python
# change the order of y axis 
plt.barh(x_axis, y_axis, label = "Boston", color = "pink")
plt.gca().invert_yaxis()

plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")

plt.legend();
```

# 3. Create Scatter Plots and Bubble Charts 

Scatter plots have many benefits over line and bar charts. A scatter plot visually depicts a positive, negative, or non-correlation relationship between two factors. 

regression analysis

```python
plt.plot(x_axis, y_axis, "o");
```

```python
plt.scatter(y_axis, x_axis, color = "red", label = "Chicago")
plt.legend()
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0, 50);
```

```python
plt.scatter(x_axis, y_axis, s = y_axis);
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

plt.scatter(x_axis, y_axis, s = y_axis_larger);

```python
#use list comprehension 
plt.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], color = "blue", linewidth = 2, label = "Boston")
plt.gca().invert_yaxis()
plt.xlim(0, 50)
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend();
```

# 4. Pie Charts

The pie charts provide a really greate way to visualize percentages. Pie charts allow us to depict percentages of categories as wedges of a pie. 

```python
plt.pie(y_axis, labels = x_axis);
```

The "explode" parameter will offset the indicated wedge by a fraction of the radius, where 0 is zero distance from the center of the pie, and 1 is completely poutside the diameter of the pie. 

```python
plt.subplots(figsize = (8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0,0,0, 0, 0)
plt.pie(y_axis, explode = explode_values, labels = x_axis, autopct = '%.1f%%',
       colors = colors);
```
